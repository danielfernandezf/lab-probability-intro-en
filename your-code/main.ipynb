## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

$$
P(\text{White}) = \frac{10}{30} = \frac{1}{3}
$$

$$
P(\text{White then Black without replacement}) = \frac{10}{30} \times \frac{8}{29} = \frac{80}{870} = \frac{8}{87}
$$

$$
P(\text{White then Black with replacement}) = \frac{10}{30} \times \frac{8}{30} = \frac{80}{900} = \frac{4}{45}
$$

$$
P(\text{3 balls all different colours}) = \frac{10}{30} \times \frac{12}{29} \times \frac{8}{28} \times 6 = \frac{48}{203}
$$

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

$$
P(\text{China}) = \frac{5}{6}, \qquad P(\text{Italy}) = \frac{1}{6}
$$

$$
P(\text{Covid} \mid \text{China}) = 0.01, \qquad P(\text{Covid} \mid \text{Italy}) = 0.20
$$

$$
P(\text{China} \mid \text{Covid}) = \frac{P(\text{Covid} \mid \text{China})P(\text{China})}{P(\text{Covid})}
$$

$$
P(\text{Covid}) = P(\text{Covid} \mid \text{China})P(\text{China}) + P(\text{Covid} \mid \text{Italy})P(\text{Italy})
$$

$$
P(\text{Covid}) = (0.01)\left(\frac{5}{6}\right) + (0.20)\left(\frac{1}{6}\right)
$$

$$
P(\text{Covid}) = \frac{0.05}{6} + \frac{0.20}{6} = \frac{0.25}{6}
$$

$$
P(\text{China} \mid \text{Covid}) = \frac{(0.01)\left(\frac{5}{6}\right)}{\frac{0.25}{6}} = \frac{0.05}{0.25} = 0.2
$$

$$
P(\text{China} \mid \text{Covid}) = 20\%
$$
